```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import seaborn as sns
import statsmodels.api as sm
```

```python
df = pd.read_csv('/content/water_consumption - water_consumption.csv')
df.head(60)
```

```python
df['block'].value_counts()
```

```python
df1 = df.copy()
df1['bloco-A'] = df1.apply(lambda row: 1 if row['block']=='A' else 0, axis=1)
df1['bloco-B'] = df1.apply(lambda row: 1 if row['block']=='B' else 0, axis=1)
df1['bloco-C'] = df1.apply(lambda row: 1 if row['block']=='C' else 0, axis=1)
df1['bloco-D'] = df1.apply(lambda row: 1 if row['block']=='D' else 0, axis=1)
df1['bloco-E'] = df1.apply(lambda row: 1 if row['block']=='E' else 0, axis=1)
df1['bloco-F'] = df1.apply(lambda row: 1 if row['block']=='F' else 0, axis=1)
df1['bloco-G'] = df1.apply(lambda row: 1 if row['block']=='G' else 0, axis=1)
df1 = df1.drop('block',axis=1)
df1.head(10)
```

```python
df1 = df1[df['price']!= 0]
df1 = df1.dropna()
df1['price'].describe()
```

```python
df1['consumption'] = df1['final'] - df1['initial']
display(df1.head())
```

```python
model = LinearRegression()
x = df1[['price','month','year','apartment','bloco-A','bloco-B','bloco-C','bloco-D','bloco-E','bloco-F','bloco-G']]
y = df1['consumption']
model.fit(x,y)
print(model.score(x, y))

x = sm.add_constant(x)
model = sm.OLS(y, x).fit()
print(model.summary())
```
